## Optimize the Model


```python
# Import our dependencies:
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv:
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
# Look at STATUS value counts:
application_df["STATUS"].value_counts()
```

```python
# Look at SPECIAL_CONSIDERATIONS value counts:
application_df["SPECIAL_CONSIDERATIONS"].value_counts()
```

```python
# Drop the non-beneficial EIN column (and also STATUS and SPECIAL_CONSIDERATIONS since they have mostly one unique value each):
application_df.drop(['EIN', 'STATUS', 'SPECIAL_CONSIDERATIONS'], axis=1, inplace=True)
application_df.head()
```

```python
# display the number of unique values for all columns:
application_df.nunique()
```

```python
# Look at NAME value counts for binning:
name_counts = application_df['NAME'].value_counts()

#  How many name counts are greater than 5?
name_counts[name_counts > 5]
```

```python
# Determine which values to replace if counts are less than or equal to 5:
names_to_replace = list(name_counts[name_counts <= 5].index)

# Replace in dataframe:
for name in names_to_replace:
  application_df['NAME'] = application_df['NAME'].replace(name, "OTHER")
    
# Check to make sure binning was successful:
application_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning:
app_type_counts = application_df['APPLICATION_TYPE'].value_counts()
app_type_counts
```

```python
# Determine which values to replace if application counts are less than 500:
app_types_to_replace = list(app_type_counts[app_type_counts < 500].index)

# Replace in dataframe:
for app_type in app_types_to_replace:
  application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app_type, "OTHER")

# Check to make sure binning was successful:
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning:
class_counts = application_df['CLASSIFICATION'].value_counts()
class_counts
```

```python
# Determine which values to replace if counts are less than 1000:
classifications_to_replace = list(class_counts[class_counts < 1000].index)

# Replace in dataframe:
for classification in classifications_to_replace:
  application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(classification, "OTHER")

# Check to make sure binning was successful:
application_df['CLASSIFICATION'].value_counts()
```

```python
# Generate our categorical variable lists:
categorical_variables = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
categorical_variables
```

```python
# perform one-hot encoding on the data:
application_with_dummies_df = pd.get_dummies(application_df)
application_with_dummies_df.head()
```

```python
# Split our preprocessed data into our features and target arrays:
X = application_with_dummies_df.drop(['IS_SUCCESSFUL'], axis=1).values
y = application_with_dummies_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset:
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Create a StandardScaler instances:
scaler = StandardScaler()

# Fit the StandardScaler:
X_scaler = scaler.fit(X_train)

# Scale the data:
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net:
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 10

nn = tf.keras.models.Sequential()

# First hidden layer:
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer:
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer:
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))

# Output layer:
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model:
nn.summary()
```

```python
# Compile the model:
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model:
nn_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data:
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {100*model_loss:.1f}%, Accuracy: {100*model_accuracy:.1f}%")
```

```python
# Export our model to HDF5 file:
nn.save("AlphabetSoupCharity_Optimization.h5")
```
